# Univariate

### Observing how data is distributed

```python
sns.histplot(x='tip', data=tips)
```

```python
sns.histplot(x='tip', data=tips,kde=True)
```

## Observation
Most of the tips are between range of 1 and 5

```python
sns.histplot(x='total_bill', data=tips)
```

```python
sns.histplot(x='total_bill', data=tips, color='red')
```

```python
sns.histplot(x='total_bill', data=tips, color='red', kde=True)
```

```python
sns.displot(x='total_bill', data=tips, color='red', kind='kde')
```

## Observation
Most of total bill is between range of 5 and 30

## Observing Outliers in Total Bill and Tips

```python
sns.boxplot(x='total_bill', data=tips)
```

```python
sns.boxplot(x='tip', data=tips)
```

## Observation
Outliers are present in both Total Bill and Tips

# Bivariate

## Observing Relationship between Total Bill and Tips

```python
sns.scatterplot(x='total_bill', y='tip', data = tips)
```

## Observation
There is a relationship between total bill and tip

```python
sns.regplot(x='total_bill', y='tip', data = tips)
```

## Observation
As the total bill increases -> tip increases

## Observing Relationship between Total Bill and Tip based on Time of Day

```python
sns.scatterplot(x='total_bill', y='tip', data = tips, hue='time')
```

```python
sns.jointplot(x='total_bill', y='tip', data = tips, hue='time')
```

## Observation
There are more visits for Dinner as compared to Lunch
Also the total bill & tips for Dinner is more as compared to Lunch

## Observing Relationship between Day and Tips based on Time of Day

```python
sns.stripplot(x='day', y='tip', hue='time', data=tips)
```

```python
sns.stripplot(x='day', y='tip', hue='time', data=tips, dodge=True)
```

```python
sns.swarmplot(x='day', y='tip', hue='time', data=tips, dodge=True)
```

## Observation
* On Sunday and Saturday -> Only tips received during Dinner
* On Thursday -> Only tips received during Lunch - except for one tip received during Dinner
* On Friday -> We have almost equal number of tips during Lunch and Dinner

## Observing Relationship between Day and Tips based on Gender

```python
sns.boxplot(x='day', y='tip', hue='gender', data=tips)
```

## Observation
* Maximum outliers on Saturday regardless of gender
* Outliers on Thursday and Friday only for Female gender

```python
sns.violinplot(x='day', y='tip', hue='gender', data=tips)
```

```python
sns.violinplot(x='day', y='tip', hue='gender', data=tips, split=True)
```

```python
sns.barplot(x='day', y='tip', hue='gender', data=tips, palette='Greens')
```

```python
sns.barplot(x='day', y='tip', hue='smoker', data=tips, palette = 'Greens')
```
